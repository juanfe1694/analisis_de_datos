```python
# Importación de paquetes y librerías
import matplotlib.pyplot as plt 
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
# Carga de información
data = pd.read_csv("C:\Analisis_de_datos\Regresion_Lineal\data.csv", sep=",")
```

```python
# Visualización de la información cargada
data
```

```python
# Visualizamos la gráfica
data.plot.scatter(x="metro", y="precio")
plt.show()
```

```python
# Agrego los datos de la variable independiente (x) y dependiente (y)
metros = data["metro"].values.reshape((-1,1))
precio = data["precio"]
```

```python
# Creo mi variable para la regresión lineal
regresion = linear_model.LinearRegression();
```

```python
# Creamos el modelo
modelo = regresion.fit(metros, precio)
```

```python
# Calculamos el intercepto (b) y la pendiente (m)
print("Interseccion (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
```

```python
# Realizamos una predicción
predicciones = modelo.predict(metros)
print(predicciones)
```

```python
# Graficamos la linea de regresión
data.plot.scatter(x="metro", y="precio", label="Datos originales")
plt.scatter(metros, predicciones, color="red", label="Predicciones")
plt.plot(metros, predicciones, color="black", label= "Linea de regresion")
plt.xlabel("Metros")
plt.ylabel("Precio")
plt.legend()
plt.show()
```

```python
# Resumen

X = sm.add_constant(metros)
model = sm.OLS(precio, X).fit()
resumen = model.summary()
print(resumen)
```

```python
# Conslusiones:
# R cuadrado: 0.737 valor de referencia 1
# p-valor (coeficiente x1): 0.141 valor de referencia 0.05
# Error estandar para el precio 42.555, para los metros 2.384
# La variabilidad de los precios respecto a los metros es de aproximadamente el 73.7%, 
# sin embargo al obtener un p-valor mayor que 0.05 se infiere que la relación entre precio y
# metros no es necesariamente significativa, lo que sugiere que el model en general no es significativo.
```
